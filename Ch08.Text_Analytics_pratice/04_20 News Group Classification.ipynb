```
# ë‚´ìš©ì„ ì œì™¸í•œ í—¤ë”ì™€ í‘¸í„°ì •ë³´ ì œê±°
# ìˆœìˆ˜íˆ ë‚´ìš©ìœ¼ë¡œ ë¶„ë¥˜í•˜ê²Œ ëª¨ë¸ í•™ìŠµì‹œí‚¤ê¸° ìœ„í•´!

from sklearn.datasets import fetch_20newsgroups

# subset='train'ìœ¼ë¡œ í•™ìŠµìš© ë°ì´í„°ë§Œ ì¶”ì¶œ, remove=('headers', 'footers', 'quotes')
train_news = fetch_20newsgroups(subset='train', remove=('header', 'footers', 'quotes'), random_state=156)
X_train = train_news.data
y_train = train_news.target
print(type(X_train))

# subset='test'ìœ¼ë¡œ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë§Œ ì¶”ì¶œ, remove=('headers', 'footers', 'quotes')
test_news = fetch_20newsgroups(subset='test',  remove=('headers', 'footers', 'quotes'), random_state=156)
X_test = test_news.data
y_test = test_news.target
print('í•™ìŠµ ë°ì´í„° í¬í‚¤ {0}, í…ŒìŠ¤íŠ¸ ë°ì´í„° í¬ê¸° {1}'.format(len(train_news.data), len(test_news.data)))
```

### **í”¼ì²˜ ë²¡í„°í™” ë³€í™˜ê³¼ ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ í•™ìŠµ/ì˜ˆì¸¡/í‰ê°€**

ğŸ’¡**ì£¼ì˜í•  ì **
* CountVectorizer.fit(X_train) -> í›„ì— CountVectorizer.transform(X_test) ìˆ˜í–‰!
* CountVectorizer.fit_transform(X_test)ì‚¬ìš©í•˜ë©´ í•™ìŠµì‹œ ì‚¬ìš©ëœ í”¼ì²˜ ê°œìˆ˜ **!=** ì˜ˆì¸¡ì‹œ ì‚¬ìš©í•  í”¼ì²˜ ê°œìˆ˜ê°€ ë˜ê¸° ë•Œë¬¸!!!

```
from sklearn.feature_extraction.text import CountVectorizer

# Count Vectorizationìœ¼ë¡œ feature extraction ë³€í™˜ ìˆ˜í–‰.
cnt_vect = CountVectorizer()
cnt_vect.fit(X_train)
X_train_cnt_vect = cnt_vect.transform(X_train)
X_test_cnt_vect = cnt_vect.transform(X_test)

print('í•™ìŠµ ë°ì´í„° Textì˜ CountVectorizer Shape:', X_train_cnt_vect.shape)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# LogisticRegressionì„ ì´ìš©í•˜ì—¬ í•™ìŠµ/ì˜ˆì¸¡/í‰ê°€ ìˆ˜í–‰.
lr_clf = LogisticRegression()
lr_clf.fit(X_train_cnt_vect, y_train)
pred = lr_clf.predict(X_test_cnt_vect)
print('CountVectorized Logistic Regressionì˜ ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorization ì ìš©í•˜ì—¬ í•™ìŠµ ë°ì´í„°ì…‹ê³¼ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ ë³€í™˜.
tfidf_vect = TfidfVectorizer()
tfidf_vect.fit(X_train)
X_train_tfidf_vect = tfidf_vect.transform(X_train)
X_test_tfidf_vect = tfidf_vect.transform(X_test)

# LogisticRegressionì„ ì´ìš©í•˜ì—¬ í•™ìŠµ/ì˜ˆì¸¡/í‰ê°€ ìˆ˜í–‰.
lr_clf = LogisticRegression()
lr_clf.fit(X_train_tfidf_vect, y_train)
pred = lr_clf.predict(X_test_tfidf_vect)
print('TF-IDF Logistic Regressionì˜ ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.3f}'.format(accuracy_score(y_test, pred)))
```

### **ìµœìƒì˜ í”¼ì²˜ ì „ì²˜ë¦¬ ìˆ˜í–‰ì„ ìœ„í•œ íŒŒë¼ë¯¸í„° íŠœë‹**

```
# stop words í•„í„°ë§ ì¶”ê°€í•˜ê³  ngramì„ ê¸°ë³¸(1, 1)ì—ì„œ (1, 2)ë¡œ ë³€ê²½í•˜ì—¬ í”¼ì²˜ ë²¡í„°í™”
tfidf_vect = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_df=300)
tfidf_vect.fit(X_train)
X_train_tfidf_vect = tfidf_vect.transform(X_train)
X_test_tfidf_vect = tfidf_vect.transform(X_test)

lr_clf = LogisticRegression()
lr_clf.fit(X_train_tfidf_vect, y_train)
pred = lr_clf.predict(X_test_tfidf_vect)
print('TF-IDF Vectorized Logistic Regression ì˜ ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.3f}'.format(accuracy_score(y_test ,pred)))
```

```
from sklearn.model_selection import GridSearchCV

# ìµœì  Cê°’ ë„ì¶œ íŠœë‹ ìˆ˜í–‰. CVëŠ” 3 Fold ì…‹ìœ¼ë¡œ ì„¤ì •.
params = {'C':[0.01, 0.1, 1 ,5, 10]}
grid_cv_lr = GridSearchCV(lr_clf, param_grid=params, cv=3, scoring='accuracy', verbose=1)
grid_cv_lr.fit(X_train_tfidf_vect, y_train)
print('Logistic Regression best C parameter :', grid_cv_lr.best_params_)

# ìµœì  Cê°’ìœ¼ë¡œ í•™ìŠµëœ grid_cvë¡œ ì˜ˆì¸¡ ìˆ˜í–‰í•˜ê³  ì •í™•ë„ í‰ê°€.
pred = grid_cv_lf.predict(X_test_tfidf_vect)
print('TF-IDF Vectorized Logistic Regressionì˜ ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.3f}'.format(accuracy_score(y_test, pred)))
```

### **ì‚¬ì´í‚·ëŸ° Pipeline ì‚¬ìš© ë° GridSearchCVì™€ì˜ ê²°í•©**

```
from sklearn.pipeline import Pipeline

# TfidfVectorizer ê°ì²´ë¥¼ tfidf_vect ê°ì²´ëª…ìœ¼ë¡œ, LogisticRegressionê°ì²´ë¥¼ lr_clfê°ì²´ëª…ìœ¼ë¡œ ìƒì„±í•˜ëŠ” Pipelineìƒì„±
pipeline = Pipeline([
                     ('tfidf_vect', TfidfVectorizer(stop_words='english',ngram_range=(1,2),max_df=300)),
                     ('lf_clf', LogisticRegression(C=10))
])

# ë³„ë„ì˜ TfidfVectorizerê°ì²´ì˜ fit_transform()ê³¼ LogisticRegressionì˜ fit(),predict()ê°€ í•„ìš” ì—†ìŒ.
# pipelineì˜ fit()ê³¼ predict()ë§Œìœ¼ë¡œ í•œêº¼ë²ˆì— í”¼ì²˜ ë²¡í„°í™”ì™€ MLí•™ìŠµ/ì˜ˆì¸¡ì´ ê°€ëŠ¥.
pipeline.fit(X_train, y_train)
pred = pipeline.predict(X_test)
print('Pipelineì„ í†µí•œ Logistic Regression ì˜ ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.3f}'.format(accuracy_score(y_test ,pred)))
```

```
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('tfidf_vect', TfidfVectorizer(stop_words='english')),
    ('lr_clf', LogisticRegression())
])

# Pipelineì— ê¸°ìˆ ëœ ê°ê°ì˜ ê°ì²´ ë³€ìˆ˜ì— ì–¸ë”ë°”(_)2ê°œë¥¼ ì—°ë‹¬ì•„ ë¶™ì—¬ GridSearchCVì— ì‚¬ìš©ë  
# íŒŒë¼ë¯¸í„°/í•˜ì´í¼ íŒŒë¼ë¯¸í„° ì´ë¦„ê³¼ ê°’ì„ ì„¤ì •.
params = {'tfidf_vect__ngram_range': [(1, 1), (1, 2), (1, 3)],
          'tfidf_vect__max_df':[100, 300, 700],
          'lr_clf__C':[1, 5, 10]}

# GridSearchCVì˜ ìƒì„±ìì— Estimatorê°€ ì•„ë‹Œ Pipeline ê°ì²´ ì…ë ¥
grid_cv_pipe = GridSearchCV(pipeline, param_grid=params, cv=3, scoring='accuracy',verbose=1)
grid_cv_pipe.fit(X_train, y_train)
print(grid_cv_pipe.best_params_, grid_cv_pipe.best_score_)

pred = grid_cv_pipe.predict(X_test)
print('Pipelineì„ í†µí•œ Logistic Regressionì˜ ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.3f}'.format(accuracy_score(y_test, pred)))
```
